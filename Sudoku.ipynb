```
data = [(1 , 1 , 8), (1 , 4, 6), (1 , 7, 9), (1 , 9, 5), (3 , 5, 2), (3 , 7 , 3), (3 , 8, 1), (4 , 3, 7), (4 , 4, 3), (4 , 5, 1), (4 , 6 , 8), (4 , 8, 6), (5 , 1, 2), (5 , 2, 4), (5 , 8, 7), (5 , 9 , 3), (7 , 3, 2), (7 , 4, 7), (7 , 5, 9), (7 , 7, 1), (8 , 1 , 5), (8 , 5, 8), (8 , 8, 3), (8 , 9, 6), (9 , 3, 3)]
```

```
# Where ever the data exists the value is made 1
for i in data:
  con = []
  row,col,val = i
  num = ((81*(row-1)) + (9*(col-1)) + val)
  con = [1 if (x == num) else 0 for x in range(1,730)]
  constraints.append(con)
```

```
# Total number of constraints
len(constraints)
```

```
# RHS of each constraint is 1
constraints_rhs = [1 for _ in range(len(constraints))]
```

```
# Range of each variable
bounds = [[0,1] for _ in range(729)]
```

```
result = linprog(c, A_eq=constraints, b_eq=constraints_rhs, bounds=bounds, method='revised simplex')
```
